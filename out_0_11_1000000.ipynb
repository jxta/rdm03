```julia
import Pkg

Pkg.add("Nemo")
Pkg.add("Hecke")
Pkg.add("PyPlot")
Pkg.add("Primes")
Pkg.add("CPUTime")

using Nemo
using Hecke
using Primes
using PyPlot: PyPlot, plt
using CPUTime

plt.figure(figsize=(30, 10))
    
P = primes(N+1000)

function plot_bsd()
    product = 1.0
    i = 3
    while P[i] < N
        p = P[i]
    
        F = GF(p)
        a = F(A)
        b = F(B)

        E = EllipticCurve([a,b], false)
        n = Int64(order_via_schoof(E))

        product *= n/p
        plt.scatter(log(p), product, s=3, color="blue")
        i = i+1
    end
end

CPUtic()

plot_bsd()

CPUtoc()

plt.xscale("log")
plt.yscale("log")
plt.show()
```
